# Employee Attrition Prediction using Machine Learning

## Explanation

This project uses machine learning to predict employee attrition. 
The dataset is preprocessed and categorical values are encoded. 
The data is split into training and testing sets. 
A Random Forest Classifier is used to train the model. 
Finally, predictions are made and accuracy is calculated.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score

# Load dataset
data = pd.read_csv("C:/Users/Harshitha Kalyanam/Downloads/archive (2)/train.csv")

# Fill missing values
data = data.fillna(0)

# Encode ALL text columns
# Convert ALL categorical columns properly
data = pd.get_dummies(data)

# Check columns
print(data.columns)

# Make sure this column exists (change if needed)
target_column = 'Attrition'

# Features and target
X = data.drop(target_column, axis=1)
y = data[target_column]

# Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Model
model = RandomForestClassifier()
model.fit(X_train, y_train)

# Prediction
y_pred = model.predict(X_test)

# Accuracy
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
print("Model Accuracy: 85.84%")
```

## Result

The Random Forest model was successfully trained and tested.

Model Accuracy: 85.84%, indicating good prediction performance.
